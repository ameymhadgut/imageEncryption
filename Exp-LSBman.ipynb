```python
from skimage import io
import numpy as np
```

```python
img = io.imread('bunny.png')
bunnyInt = np.array(img)
bunny = bunnyInt.astype(float)
bunny /= 255
io.imshow(bunny)
img = io.imread('cube.png')
cubeInt = np.array(img)
cube = cubeInt.astype(float)
cube /= 255
print(cube[450][600])
enc_param = 1000
```

```python
io.imshow(cube)
```

```python
hidden = []
for i in range(len(bunny)):
    hidden_row = []
    for j in range(len(bunny[i])):
        if(bunnyInt[i][j].any()<0.9):
            hidden_row.append(bunnyInt[i][j]+cubeInt[i][j]//enc_param)
        else:
            hidden_row.append(bunnyInt[i][j]-cubeInt[i][j]//enc_param)
    hidden.append(hidden_row)
print(hidden[450][600])
io.imshow(np.array(hidden))
hidden_2 = np.array(hidden) *255
test = hidden_2.astype('uint8')
io.manage_plugins.call_plugin('imsave', 'new_test1.png', test)
io.imsave('new_test2.png',np.array(hidden))
```

```python
img = io.imread('new_test1.png')
hiddenInt = np.array(img)
hidden = hiddenInt.astype(float)
hidden /= 255
io.imshow(hidden)
print(hidden[450][600])
```

```python
enc = []
for i in range(len(hiddenInt)):
    enc_row = []
    for j in range(len(hiddenInt[i])):
        if(bunny[i][j].any()<0.9):
            enc_row.append((hiddenInt[i][j] - bunnyInt[i][j])*enc_param)
        else:
            enc_row.append((bunnyInt[i][j]- hiddenInt[i][j])*enc_param)
    enc.append(enc_row)
io.imshow(np.array(enc))
print(enc[200][1000])
```
