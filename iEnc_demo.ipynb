```python
from iEnc import encrypt, decrypt, rgb_to_grey, get_loss
from skimage import io
```

```python
print("Place holder image")
io.imshow('bunny.png')
```

```python
print("Image to be hidden")
io.imshow('cube.png')
```

```python
print("Encrypted image using dct")
encrypt('bunny.png','cube.png','dct','out.png',shuffle=1,r1=10,r2=20)
io.imshow('out.png')
```

```python
print("Decrypted hidden image using dct")
decrypt('bunny.png','out.png','dct','out1.png')
io.imshow('out1.png')
print(get_loss('cube.png', 'out1.png'))
```

```python
print("Decrypted hidden image using dct")
decrypt('bunny.png','out.png','dct','out1.png',shuffle=1,r1=10,r2=3)
io.imshow('out1.png')
print(get_loss('cube.png', 'out1.png'))
```

```python
print("Decrypted hidden image using dct")
decrypt('bunny.png','out.png','dct','out1.png',shuffle=1,r1=4,r2=20)
io.imshow('out1.png')
print(get_loss('cube.png', 'out1.png'))
```

```python
print("Decrypted hidden image using dct")
decrypt('bunny.png','out.png','dct','out1.png',shuffle=1,r1=4,r2=3)
io.imshow('out1.png')
print(get_loss('cube.png', 'out1.png'))
```

```python
print("Decrypted hidden image using dct")
decrypt('bunny.png','out.png','dct','out1.png',shuffle=1,r1=10,r2=20)
io.imshow('out1.png')
print(get_loss('cube.png', 'out1.png'))
```
