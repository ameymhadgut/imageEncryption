```python
from iEnc import encrypt, decrypt, rgb_to_grey
from skimage import io
```

```python
print("Place holder image")
io.imshow('bunny.png')
```

```python
print("Image to be hidden")
io.imshow('cube.png')
```

```python
print("Encrypted image using dct")
encrypt('bunny.png','cube.png','dct','out.png')
io.imshow('out.png')
```

```python
print("Decrypted hidden image using dct")
decrypt('bunny.png','out.png','dct','out1.png')
io.imshow('out1.png')
```

```python
print("Place holder image grayscale")
io.imshow(rgb_to_grey(io.imread('bunny.png')))
```

```python
print("Grayscale Image to be hidden")
io.imshow(rgb_to_grey(io.imread('cube.png')))
```

```python
print("Encrypted image using dctdwt")
encrypt('bunny.png','cube.png','dctdwt','out2.png')
io.imshow('out2.png')
```

```python
print("Decrypted image using dctdwt")
decrypt('bunny.png','out2.png','dctdwt','out3.png')
io.imshow('out3.png')
```
